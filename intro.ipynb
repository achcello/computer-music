For now, all the parsing will be done manually while I attempt to figure out music21.

```python
# parse the files into score objects and set the tempi for the pieces
sample1 = converter.parse('./Sample_1.xml')
sample2 = converter.parse('./Sample_2.xml')
```

```python
sample1.plot('pitch')
analysis.discrete.analyzeStream(sample1, 'key')
```

```python
sample2.plot('pitch')
analysis.discrete.analyzeStream(sample2, 'key')
```

This is how to receive pitches from the score object which was interpreted.

```python
sample1Pitches = sample1.parts[0].pitches
sample1PitchesList = [str(p) for p in sample1Pitches]
sample1Freqs = [p.midi for p in sample1Pitches]
print(sample1PitchesList, "\n")
print(sample1Freqs)
```

```python
sample2Pitches = sample2.parts[0].pitches
sample2PitchesList = [str(p) for p in sample2Pitches]
sample2Freqs = [p.midi for p in sample2Pitches]
print(sample2PitchesList, "\n")
print(sample2Freqs)
```

A possible metric could be a kind of 'rate' which signifies notes per second

```python
sample1Rate = len(sample1Pitches) / 8 #seconds
print(sample1Rate)
```

```python
sample2Rate = len(sample2Pitches) / 16 #seconds
print(sample2Rate)
```
